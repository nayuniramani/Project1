```python
## Aprior algorithm
```

```python
# for support is 0.2
frequent_itemsets = apriori(data, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
# for support is 0.3
frequent_itemsets = apriori(data, min_support=0.3, use_colnames=True)
frequent_itemsets
```

```python
# for confidence 0.6
rules1 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
rules1
rules1.sort_values('lift',ascending = False)
```

```python
# for confidence 0.7
rules2 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules2
rules2.sort_values('lift',ascending = False)
```
