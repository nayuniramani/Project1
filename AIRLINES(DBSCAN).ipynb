```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/DELL/Desktop/EastWestAirlines.csv")
```

```python
data.info()
```

```python
data.drop(['ID#'],axis=1,inplace=True)
```

```python
data.info()
```

```python
# Normalizing the data
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(data)
```

```python
df_norm.head()
```

```python
# for eplision value 1 it gives 0,1 clusters
dbscan = DBSCAN(eps=1, min_samples=11)   
dbscan.fit(df_norm)
```

```python
set(dbscan.labels_)
```

```python
## for epsilon value is 1.5 it gives only one cluster
dbscan = DBSCAN(eps=1.5, min_samples=11)   
dbscan.fit(df_norm)
```

```python
set(dbscan.labels_)
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
clustered = pd.concat([data,cl],axis=1)
```
